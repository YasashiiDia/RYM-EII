```
#@title Pie Charts
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive_output
%matplotlib inline
sns.set()

mpl.rcParams['font.size'] = 20.0
labelfontsize = 14
textprops={'fontsize': 14, 'family':'DejaVu Sans Mono'}
textcolor = 'dimgray'

# Define rectangle in axes coords
left, width = .25, .5
bottom, height = .25, .5
right = left + width
top = bottom + height

bboxprops = dict(boxstyle='round', pad=0.5, alpha=0.8, facecolor='lemonchiffon', edgecolor='darkgoldenrod')

# Slice pie in accetable sizes and put remaining pie in 'Other'
def pie_slicer(series, cm=None, thresh=0.02):
  tot = series.sum()
  other, n_slices = 0, 0
  serieslist = series.tolist()
  labels, sizes = [], []

  for i, s in enumerate(series):
    if s/tot < thresh:
      other += s
    else:
      n_slices += 1
      labels.append(series.index[i])
      sizes.append(s)
  
  if other/tot > thresh:
    labels.append('Other')
    sizes.append(other)

  colors = [plt.get_cmap('Pastel1')(i) for i in range(9)]
  [colors.append(plt.get_cmap('Pastel2')(i)) for i in range(9)]
  if cm:
    colors = [plt.get_cmap(cm)(i) for i in np.linspace(0, 1, n_slices+2)]
  return labels, sizes, colors

# Nicely format text for textbox (must use mono font for proper alignment)
def box_format(series,n):
    longest = series.index.str.len().max()
    whitespaces = [''.rjust(longest-len(s)+1) if series[s] < 10 else ''.rjust(longest-len(s)) for s in series.index.values]
    t = [str(series.index.values[i])+':'+whitespaces[i]+str(series.values[i])+'\n' for i in range(len(series))]
    text = 'Top '+ str(n) +' Actors\n\n'+''.join(t)[:-1]
    return text

def pies(Top, Year, Country, ExCountry, Gender, InGenres, ExGenres, GroupRank, Age, Weight, Sort, Ascending, Print):

    try:
      a = filt(df, 1001, Year, Country, ExCountry, Gender, InGenres, ExGenres, GroupRank, Age, Sort, Ascending)    
      a = a[(a['Rank'] <= Top[1]) & (a['Rank'] >= Top[0])]

      fig, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize = (18,6))
      plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.4, hspace=0.6)

      # Country pie
      if Weight == 'None':
        g = a['Place of Birth']
        g = g.groupby(g).size().sort_values(ascending=False)
      else:
        g = a.groupby(['Place of Birth'])[Weight].sum().sort_values(ascending=False)

      labels, sizes, colors = pie_slicer(g)

      patches, texts, autotexts = ax2.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90,textprops=textprops)
      ax2.axis('equal')
      [text.set_color(textcolor) for text in texts]
      [autotext.set_color(textcolor) for autotext in autotexts]

      # Genre pie
      expl = a
      expl['Genres'] = expl['Genres'].str.split(', ')
      expl = expl.explode('Genres')
      if Weight == 'None':
        g = expl['Genres']
        g = g.groupby(g).size().sort_values(ascending=False)
      else:
        g = expl.groupby('Genres')[Weight].sum().sort_values(ascending=False)

      labels, sizes, colors = pie_slicer(g)

      patches, texts, autotexts = ax1.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90,textprops=textprops)
      ax1.axis('equal')
      [text.set_color(textcolor) for text in texts]
      [autotext.set_color(textcolor) for autotext in autotexts]

      # Gender pie
      if Weight == 'None':
        g = a['Gender']
        g = g.groupby(g).size().sort_values(ascending=False)
      else:
        g = a.groupby(['Gender'])[Weight].sum().sort_values(ascending=False)

      labels, sizes, colors = pie_slicer(g)
      if Gender == 'Male':
        labels = ['Male']
      elif Gender == 'Female':
        labels = ['Female']
      else:
        labels = ['Male','Female']
        sizes=sizes[:2]

      patches, texts, autotexts = ax3.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90,textprops=textprops)
      ax3.axis('equal')

      [text.set_color(textcolor) for text in texts]
      [autotext.set_color(textcolor) for autotext in autotexts]

      # Top Actors
      '''topactors = a.groupby('Name')['Name'].count()
      n = min(15,topactors.size)
      topactors = topactors.sort_values(ascending=False).iloc[:n]
      text = box_format(topactors,n)

      ax3.text(0.5*(left+right), 0.5*(bottom+top), text, family='DejaVu Sans Mono',
          horizontalalignment='center', verticalalignment='center',
          fontsize=15, color='dimgrey', transform=ax3.transAxes, bbox=bboxprops)
      ax3.axis('off')'''

      plt.tight_layout()
      plt.show()

    except ValueError:
      print('\nNo entries match criteria')
      pass

# Widgets
n_titles=len(df)
max_age = df['Age'].max()
layout={'width': '350px'}

top_wp = widgets.IntRangeSlider(min=1, max=n_titles, step=1,value=(1,n_titles),layout=layout,description='Rank:',continuous_update=False)
weight_wp = widgets.Dropdown(options=['None','Points'],layout=layout,value='None',description='Weight:')

year_wp = widgets.IntRangeSlider(min=1917, max=2019, step=1,layout=layout,value=[1917,2019],description='Year:')
sort_wp = widgets.Dropdown(options=['Rank','Release Date','Age','Group','Group Rank'],layout=layout,value='Rank',description='Sort:')
gender_wp = widgets.Dropdown(options=['All','Male','Female'],layout=layout,value='All',description='Gender:')
check_wp = widgets.Checkbox(value=True,description='Ascending')
print_wp = widgets.Checkbox(value=False,description='Forum Print')

group_wp = widgets.IntRangeSlider(min=1, max=51, step=1,layout=layout,value=[1,51],description='Gr. Rank:')
age_wp = widgets.IntRangeSlider(min=0, max=max_age, step=1,layout=layout,value=[0,max_age],description='Age:')

exgenres_wp = widgets.Combobox(layout=layout,options=unique_genres,placeholder='Exclude Genre',description='Ex. Genre:')
ingenres_wp = widgets.Combobox(layout=layout,options=unique_genres,placeholder='Include Genre',description='In. Genre:')

country_wp = widgets.Combobox(layout=layout,options=unique_countries,placeholder='Include Place of Birth',description='In. Country:')
excountry_wp = widgets.Combobox(layout=layout,options=unique_countries,placeholder='Exclude Place of Birth',description='Ex. Country:')


wsp = [top_wp,year_wp,group_wp,age_wp,country_wp,excountry_wp,ingenres_wp,exgenres_wp,gender_wp, weight_wp]

interactive_pies = interactive_output(pies, {'Top':top_wp,'Year':year_wp,'Country':country_wp,'ExCountry':excountry_wp,'Gender':gender_wp,'InGenres':ingenres_wp,'ExGenres':exgenres_wp,'Ascending':check_wp,'Sort':sort_wp,'Print':print_wp,'GroupRank':group_wp,'Age':age_wp,'Weight':weight_wp})
interactive_pies.layout.height = '350px'

ui_pies = widgets.GridBox(wsp, layout=widgets.Layout(grid_template_columns="repeat(2, 400px)"))

display(ui_pies, interactive_pies)
```
